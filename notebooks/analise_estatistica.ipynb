# Análise dos Microdados da RAIS (2021–2024)

Este trabalho apresenta uma análise exploratória dos **microdados da RAIS** referentes aos trabalhadores com vínculos ativos no período de **2021 a 2024**. O objetivo é compreender o comportamento do mercado de trabalho formal brasileiro, destacando características dos trabalhadores e evolução ao longo dos anos.


**Instalação de Pacotes Necessários**

O script abaixo tem como objetivo **garantir que todos os pacotes necessários para o projeto estejam instalados** no ambiente R.  

A lista de pacotes inclui ferramentas para **ciência de dados, visualização, manipulação de grandes bases, geração de relatórios e tabelas interativas**. Entre eles:  
- `tidyverse` → coleção de pacotes como `dplyr`, `ggplot2`, `tidyr`;  
- `data.table` → manipulação eficiente de grandes conjuntos de dados;  
- `rmarkdown` → criação de relatórios e notebooks;  
- `here` → gerenciamento de caminhos de arquivos;  
- `progress` → exibição de barras de progresso;  
- `beepr` → notificações sonoras;  
- `moments` → cálculo de momentos estatísticos (assimetria e curtose);  
- `gridExtra` → combinação de múltiplos gráficos;  
- `DT` → geração de tabelas interativas;  
- `kableExtra` → formatação avançada de tabelas em HTML e PDF.  

O código verifica quais pacotes **ainda não estão instalados** e instala somente os que faltam, evitando reinstalações desnecessárias.

**Estatística Descritiva — RAIS (2021–2024)**

O script realiza **análises descritivas detalhadas** para cada ano da RAIS, considerando apenas vínculos ativos (`Vínculo Ativo 31/12 == 1`).  

Para as variáveis **Idade**, **Horas Contratuais** e **Remuneração Média**, são calculadas as seguintes métricas:  
- Média e mediana  
- Desvio-padrão  
- Mínimo e máximo  
- 1º quartil (P25) e 3º quartil (P75)  
- Assimetria (skewness)  
- Curtose (kurtosis)

O código percorre os arquivos de 2021 a 2024, carrega cada base e imprime a **tabela de estatísticas descritivas**, permitindo comparar a distribuição e a variabilidade dessas variáveis ao longo dos anos.


```r
source("c:/analise_estatistica/src/estatistica_descritiva.R")
```

**Visualizações — RAIS Paraíba (2021–2024)**

O script gera gráficos exploratórios para analisar o perfil dos trabalhadores com **vínculos ativos** entre 2021 e 2024. As bases são preparadas e combinadas em uma tabela única (`dados_todos`) para facilitar a visualização comparativa.  

As análises gráficas incluem:

1. **Histogramas da idade (2x2)**  
   - Mostram a distribuição etária dos trabalhadores por ano.

2. **Boxplots da remuneração (2x2)**  
   - Com limite superior no percentil 95 para reduzir o efeito de outliers, permitindo comparar a dispersão salarial anual.

3. **Linha da média salarial (único gráfico)**  
   - Mostra a evolução da média de remuneração de 2021 a 2024.

4. **Gráficos de barras da escolaridade (2x2)**  
   - Percentual de trabalhadores por nível de escolaridade em cada ano.

5. **Gráficos de barras por sexo (2x2)**  
   - Percentual de homens e mulheres entre os trabalhadores ativos por ano.

Todas as visualizações utilizam o **tema minimalista do ggplot2** e são dispostas em **grades 2x2** usando `gridExtra::grid.arrange` para facilitar a comparação entre anos.


```r
source("c:/analise_estatistica/src/graficos.R")
```
